>>**BEV (Battery Electric Vehicle)**: Fully electric vehicles powered only by batteries with no internal combustion engine.

###<u>The dataset includes multiple metrics related to electric vehicles:</u>

>>**EV sales (number of vehicles)**: The total count of electric vehicles sold within a region and year.

>>**EV stock share (percentage)**: The proportion of electric vehicles within the overall vehicle fleet, indicating market penetration of EVs in the total stock.

>>**EV sales share (percentage)**: The share of electric vehicles sold as a percentage of all vehicle sales, reflecting how popular EVs are among new car buyers.

# **Investigating Representation of Countries in EV Sales and Growth Rates in Sales**


---



While invesitagting the EV sales in each country from 2010 to 2023, we assumed that the highest total sales for those years would come from the countries with the highest GDP's. To test this hypothesis, we performed a two-sample t-test comparing the total sales from 2010 to 2023 for the 5 countries with the highest GDP's: USA, China, Germany, India, and Japan to the total sales from the other countries. By seeing if the difference between these two samples was significant, we could check to see if our assumption was correct.

After analyzing total EV sales, we looked at the growth rate in sales from 2010 to 2023. Using the calculated rate, we predicted the total EV in 2023 for each country in the dataset, based on data from the previous year. Our hypothesis is that the there is no significant difference between the predicted and actual EV sales values in 2023. This assumes that, on average, the countries have relatively consistent, compouding growth rates .

```python
# summing ev sales over time for each country
counts = []
for country in sales_df['region'].unique():
  counts.append(sales_df[sales_df['region'] == country]['value'].sum())
# plotting ev sales against country
plt.figure(figsize=(10, 10))
plt.barh(sales_df['region'].unique(), counts, color = 'violet')
plt.title('Frequency of Countries')
plt.xlabel('EV Car Sales (10M)')
plt.ylabel('Country')
plt.show()
# summing ev sales for top 5 GDP countries
topgppcountries = ["USA", "China", "Germany", "India", "Japan"]
topgdpcounts = []
for country in topgppcountries:
  topgdpcounts.append(sales_df[sales_df['region'] == country]['value'].sum())
# summing ev sales for other countries
othercounts = []
for country in sales_df['region'].unique():
  if country not in topgppcountries:
    othercounts.append(sales_df[sales_df['region'] == country]['value'].sum())
# finding out if difference is significant by performing two sample T-test
# no reason to assume variances are equal, so Welch's test
t_stat, p_value = ttest_ind(topgdpcounts, othercounts, equal_var=False, alternative='greater')
print("T-statistic:", t_stat)
print("P-value:", p_value)
plt.figure(figsize=(10, 10))
grouped = sales_df.groupby(['region', 'year'], as_index=False)['value'].sum()
grouped = grouped.sort_values(by = ['region', 'year'])
yeardiff = grouped.groupby('region')['year'].max() - grouped.groupby('region')['year'].min()
first_val = grouped.groupby('region')['value'].first()
last_val = grouped.groupby('region')['value'].last()
growth_rates = (last_val / first_val)**(1 / yeardiff) - 1
plt.figure(figsize=(10, 10))
plt.barh(growth_rates.index, growth_rates.values*100, color = 'blue')
plt.title('Growth Rates of Countries')
plt.xlabel('Growth Rate (%)')
plt.ylabel('Country')
plt.show()
# predict sales in 2023
predicted_sales = []
for country in sales_df['region'].unique():
  predicted_sales.append(growth_rates[country] * sales_df[(sales_df['region'] == country) & (sales_df['year'] == 2022)]['value'].sum() + sales_df[(sales_df['region'] == country) & (sales_df['year'] == 2022)]['value'].sum())
# actual sales in 2023
actual_sales = []
for country in topgppcountries:
  actual_sales.append(sales_df[(sales_df['region'] == country) & (sales_df['year'] == 2023)]['value'].sum())
# finding out if difference is significant
# we can assume equal variance since data is spread approximately the same
t_stat2, p_value2 = ttest_ind(actual_sales, predicted_sales, equal_var=True, alternative='greater')
print("T-statistic:", t_stat2)
print("P-value:", p_value2)
```

# *Correlation of EV Type Sales Growth*

While the format of our cleaned sales_df dataset does not show the correlation between the sales growth of different EV types, reorganizing the dataframe allows us to see how these sales are correlated through a correlation matrix.

We can assume the null hypothesis to be the assumption that there is a strong positive correlation between all EV types, as an increase in EV sales in general may increase sales for all types of EVs, presumably as a similar rate.

Thus, our alternative hypothesis is that the sales of at least one pair of EV types do not have a strong poitive correlation, and that our assumed trend of EV sales types is incorrect.

```python
type_cor_df = sales_df.copy()
type_cor_result = type_cor_df.pivot_table(
    index=['region', 'year'],
    columns='powertrain',
    values='value',
    fill_value = 0
).reset_index()
type_cor_result = type_cor_result[['region', 'BEV', 'PHEV', 'FCEV', 'year']]

correlation_matrix = type_cor_result[['BEV', 'PHEV', 'FCEV']].corr()
correlation_matrix
```

There appears to be a strong correlation between BEV and PHEV type sales worldwide. BEV and FCEV have a minimal positive correlation, with PHEV and FCEV having practically no positive correlation. As a result, our alternative hypothesis was proven correct-- not all EV types have a distinct positive correlation.

#**Exploring Volatility Calculation and Visualization**

```python
ev_agg['pct_change'] = ev_agg.groupby('region')['ev_sales'].pct_change()

volatility = ev_agg.groupby('region')['pct_change'].std().reset_index()
volatility.columns = ['region', 'volatility']
volatility = volatility.sort_values(by='volatility', ascending=False)

plt.figure(figsize=(12, 6))

volatility_sorted = volatility.sort_values('volatility', ascending=True)

plt.barh(
    volatility_sorted['region'],
    volatility_sorted['volatility'],
    color='skyblue'
)
plt.xlabel('Standard Deviation of YoY Growth (%)')
plt.title('EV Sales Growth Volatility (2010–2023) in Top 10 GDP Countries')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

**Volatility** in this context refers to the **consistency (or inconsistency)** in a country's year-over-year (YoY) electric vehicle (EV) sales growth. It's measured using the **standard deviation** of the annual percentage change in EV sales from 2010 to 2023.

*   A **higher standard deviation** means the country's EV sales growth has fluctuated more, indicating an **unstable growth pattern.**

*   A **lower standard deviation** suggests that the country has experienced **steadier growth** year to year.

This horizontal bar chart displays the **volatility of EV sales growth** across the **top 10 GDP countries.** Here’s what we can observe:
The United States has the highest volatility, meaning its annual EV growth has

*   The **United States** has the **highest volatility**, meaning its annual EV growth has varied widely — possibly due to shifts in policy, consumer demand, or manufacturer output.

*   **Germany** and **France** also show relatively high volatility, reflecting some inconsistency in how their EV markets have grown over the years.

*   On the other end, **Canada, Italy**, and **China** show the **lowest volatility**, suggesting that their EV markets have grown at a more consistent pace.




```python
from scipy.stats import f_oneway

ev_agg['pct_change'] = ev_agg.groupby('region')['ev_sales'].pct_change()

ev_agg_top10 = ev_agg[ev_agg['region'].isin(top10_gdp)]

anova_groups = [ev_agg_top10[ev_agg_top10['region'] == country]['pct_change'].dropna()
                for country in top10_gdp]

f_stat, p_value = f_oneway(*anova_groups)
print("F-statistic:", f_stat)
print("P-value:", p_value)
```

The F-statistic is very low, and the corresponding p-value is substantially higher than the common significance threshold (e.g., 0.05). This indicates no statistically significant difference in the mean year-over-year EV sales growth rates among the top 10 GDP countries.

In simpler terms, while the earlier volatility analysis (using standard deviation) showed that some countries have more fluctuating growth patterns than others, the overall average percentage growth rates are statistically similar across these countries over the period from 2010 to 2023.

Thus, although the magnitude of fluctuations (volatility) varies, the ANOVA suggests there is no evidence to conclude that the average YoY growth rates are fundamentally different among these nations.

*   This setup allows us to identify both global trends and local adoption patterns, offering a data-driven view into how EV markets are expected to evolve.

# **Regression Model - Countries**

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd

top10_gdp = [
    'USA', 'China', 'Germany', 'India', 'Japan',
    'United Kingdom', 'France', 'Italy', 'Canada', 'Brazil'
]

power_trains = ['BEV', 'PHEV', 'FCEV']
years = list(range(2010, 2024))  # 2010–2023

for country in top10_gdp:
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))
    fig.suptitle(country, fontsize=20, y=1.05)
    predictions_2024 = []

    for i, power in enumerate(power_trains):
        # Filter original data
        df = sales_df[(sales_df['region'] == country) & (sales_df['powertrain'] == power)]

        # Create full year range with zero or NaN if missing
        full_years_df = pd.DataFrame({'year': years})
        merged_df = pd.merge(full_years_df, df[['year', 'value']], on='year', how='left')
        merged_df['value'] = merged_df['value'].fillna(0)  # or use interpolation if preferred

        x = merged_df[['year']]
        y = merged_df['value']

        model = LinearRegression()
        model.fit(x, y)
        predictions = model.predict(x)

        # Predict for 2024
        prediction_2024 = model.predict(pd.DataFrame({'year': [2024]}))[0]
        predictions_2024.append(prediction_2024)

        # Plot
        axes[i].scatter(x, y, color='orange')
        axes[i].plot(x, predictions, color='blue')
        axes[i].set_title(power)
        axes[i].set_xlabel('YEAR')
        axes[i].set_ylabel('EV SALES')

    plt.tight_layout()
    plt.show()

    print("2024 Predictions for", country)
    for i in range(3):
        print(f"{power_trains[i]}: {predictions_2024[i]:.2f}")




```

# **Regression Model - Global Scale**

This code performs a global analysis of electric vehicle (EV) sales by aggregating the dataset across all countries based on year and powertrain type. It focuses on three major EV categories: BEV (Battery Electric Vehicles), PHEV (Plug-in Hybrid Electric Vehicles), and FCEV (Fuel Cell Electric Vehicles). For each powertrain, a linear regression model is trained using historical worldwide sales data to identify trends and forecast sales for the year 2024. The model takes the year as the input feature and total global EV sales as the target variable. The results are visualized using three side-by-side subplots, each showing the actual data points (in yellow) and the fitted regression line (in blue) for one powertrain type. After the plots are displayed, the code prints the predicted sales figures for 2024, offering a simple yet insightful projection of global EV market growth by technology type.

**Modeling Approach**

The modeling approach used in this analysis is based on linear regression, a supervised machine learning technique that models the relationship between a continuous dependent variable and one or more independent variables. In this case, the independent variable is the year, and the dependent variable is the total EV sales for each powertrain type. For each powertrain (BEV, PHEV, FCEV), a separate linear model is trained on the aggregated historical global sales data. The assumption is that past trends in EV adoption can be approximated by a linear growth pattern over time. Once trained, each model is used to predict the expected sales for the year 2024. This approach provides a straightforward and interpretable way to forecast future sales, serving as a baseline for understanding EV adoption trends globally.

```python
grouped = sales_df.groupby(['year', 'powertrain'])['value'].sum().reset_index()
power_trains = ['BEV', 'PHEV', 'FCEV']

fig, axes = plt.subplots(1, 3, figsize=(15, 4))  # 1 row, 3 columns
fig.suptitle('world', fontsize = 20, y = 1.05)
predictions_2024 = []
for i in range(len(power_trains)):
  power = power_trains[i]
  df = grouped[grouped['powertrain']==power]
  if len(df)==0:
    continue
  x = df[['year']]
  y = df['value']
  model = LinearRegression()
  model.fit(x,y)
  predictions = model.predict(x)
  new_data = pd.DataFrame({'year': [2024]})
  predictions_2024.append(model.predict(new_data))
  axes[i].scatter(x,y,color='orange')
  axes[i].plot(x, predictions,color= 'blue')
  axes[i].set_title(power)
  axes[i].set_xlabel('YEAR')
  axes[i].set_ylabel('EV SALES')
plt.show()
print("2024 Predictions")
for i in range(3):
  print(power_trains[i] + ": "+str(predictions_2024[i]))



```

# **Attempting Additional Regressions**

While the model finds and fits a linear equation to the given datasets, it cannot be said that the fit is ideal. Thus, we will attempt to fit two other regression types -- polynomial and exponential -- to see which best fits each graph, mainly through each line's estimate of sales in 2023. Then, the best fitting model will predict the number of recorded sales in 2024.

Our code will print the overlapping graphs, the best fit regression and the 2023 squared error.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

top10_gdp = ['USA', 'China', 'Germany', 'India', 'Japan', 'United Kingdom', 'France', 'Italy', 'Canada', 'Brazil']
power_trains = ['BEV', 'PHEV', 'FCEV']

for country in top10_gdp:
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle(country, fontsize=20, y=1.08)
    for i, power in enumerate(power_trains):
        df = sales_df[(sales_df['region'] == country) & (sales_df['powertrain'] == power)]

        # Ensure data starts from 2010 to 2023 (training data only)
        full_years = pd.DataFrame({'year': list(range(2010, 2024))})
        df = pd.merge(full_years, df[['year', 'value']], on='year', how='left')
        df['value'] = df['value'].fillna(0)  # or use interpolate() if you prefer

        df_fit = df[df['year'] <= 2022]
        df_2023 = df[df['year'] == 2023]
        if len(df_fit) == 0 or len(df_2023) == 0:
            axes[i].set_title(f"{power}\nNo Data")
            axes[i].axis('off')
            continue
        x = df_fit[['year']]
        y = df_fit['value'].values
        x_plot = pd.DataFrame({'year': np.arange(x['year'].min(), 2024)})
        # Linear
        lin_model = LinearRegression()
        lin_model.fit(x, y)
        lin_pred_2023 = lin_model.predict(pd.DataFrame({'year': [2023]}))[0]
        lin_fit = lin_model.predict(x_plot)
        # Polynomial (degree=2)
        poly = PolynomialFeatures(degree=2, include_bias=False)
        x_poly = poly.fit_transform(x)
        poly_model = LinearRegression()
        poly_model.fit(x_poly, y)
        poly_pred_2023 = poly_model.predict(poly.transform(pd.DataFrame({'year': [2023]})))[0]
        poly_fit = poly_model.predict(poly.transform(x_plot))
        # Exponential
        log_y = np.log(y+1)
        exp_model = LinearRegression()
        exp_model.fit(x, log_y)
        exp_pred_2023 = np.exp(exp_model.predict(pd.DataFrame({'year': [2023]})))[0]
        exp_fit = np.exp(exp_model.predict(x_plot))

        actual_2023 = df_2023['value'].values[0]
        # Error (MSE)
        errors = {'Linear': abs(lin_pred_2023 - actual_2023) ** 2,
                  'Polynomial': abs(poly_pred_2023 - actual_2023) ** 2,
                  'Exponential': abs(exp_pred_2023 - actual_2023) ** 2}
        percent_errors = {'Linear': abs(lin_pred_2023 - actual_2023) / abs(actual_2023) * 100,
                          'Polynomial': abs(poly_pred_2023 - actual_2023) / abs(actual_2023) * 100,
                          'Exponential': abs(exp_pred_2023 - actual_2023) / abs(actual_2023) * 100}
        best_fit = min(errors, key=errors.get)

        axes[i].scatter(x, y, color='orange', label='Train Data')
        axes[i].scatter([2023], [actual_2023], color='red', marker='x', s=100, label='Actual 2023')
        axes[i].plot(x_plot, lin_fit, color='blue', label='Linear')
        axes[i].plot(x_plot, poly_fit, color='green', label='Polynomial')
        axes[i].plot(x_plot, exp_fit, color='purple', label='Exponential')
        axes[i].scatter([2023], [lin_pred_2023], color='blue', marker='o', label='Linear Fit')
        axes[i].scatter([2023], [poly_pred_2023], color='green', marker='o', label='Polynomial Fit')
        axes[i].scatter([2023], [exp_pred_2023], color='purple', marker='o', label='Exponential Fit')
        axes[i].set_title(f"{power}\nBest: {best_fit}")
        axes[i].set_xlabel('YEAR')
        axes[i].set_ylabel('EV SALES')
        axes[i].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x):,}'))
        axes[i].legend(fontsize=8)
        # 2024 Prediction
        if best_fit == 'Linear':
            pred_2024 = lin_model.predict(pd.DataFrame({'year': [2024]}))[0]
        elif best_fit == 'Polynomial':
            pred_2024 = poly_model.predict(poly.transform(pd.DataFrame({'year': [2024]})))[0]
        elif best_fit == 'Exponential':
            pred_2024 = np.exp(exp_model.predict(pd.DataFrame({'year': [2024]})))[0]
        print(f"{country} - {power}: Best Fit = {best_fit}, \n     "
              f"2023 MSE = {errors[best_fit]:,.0f} "
              #f"2023 Percent Error = {percent_errors[best_fit]:.2f}%, \n     "
              #f"2024 {best_fit} Prediction: {pred_2024:,.0f} {power} Sales")
        )
    plt.show()
```

Despite the exponential-like trend at which the sales of these countries grow, the best fit is often the polynomial line. So the distribution of points is neither a clear linear of exponential set of points. It can be extrapolated, then, that the sales of EV cars in the top 10 highest GDP countries will continuously grow at this polynomial rate. Because polynomial regression often fits datasets that have distinct changes in growth rate, this implies that the growth rate of these countries' sales may increase further in the future, due to the rapid increase around the 2020.

# 🧪 **Model Accuracy Check: 2023 Predictions vs Actuals**

In this section, we tested how well our polynomial regression model performs by predicting EV sales for 2023 and comparing those predictions to the actual reported values. We trained the model using data up to 2022, then made predictions for 2023 and calculated the percent error for each country and powertrain.

```python
for country in top10_gdp:
    print(f"\n--- {country} ---")
    for power in power_trains:
        df = sales_df[(sales_df['region'] == country) & (sales_df['powertrain'] == power)]

        if 2023 not in df['year'].values or len(df) < 5:
            continue

        train = df[df['year'] < 2023]
        test_2023 = df[df['year'] == 2023]['value'].values[0]

        x_train = train[['year']]
        y_train = np.log(train['value'].values)

        poly = PolynomialFeatures(degree=2, include_bias=False)
        x_train_poly = poly.fit_transform(x_train)

        model = LinearRegression()
        model.fit(x_train_poly, y_train)

        x_2023_df = pd.DataFrame({'year': [2023]})
        x_2024_df = pd.DataFrame({'year': [2024]})

        x_2023 = poly.transform(x_2023_df)
        x_2024 = poly.transform(x_2024_df)

        pred_2023_log = model.predict(x_2023)[0]
        pred_2024_log = model.predict(x_2024)[0]

        pred_2023 = np.exp(pred_2023_log)
        pred_2024 = np.exp(pred_2024_log)

        error_pct = abs(pred_2023 - test_2023) / test_2023 * 100

        print(f"{power} | 2023 predicted: {int(pred_2023):,}, actual: {int(test_2023):,}, error: {error_pct:.2f}%")
        print(f"{power} | 2024 forecast: {int(pred_2024):,}")
```

The results showed that while the model captures general trends, it struggles with precision for some countries and powertrains—especially where the sales numbers change rapidly or the data is limited. However, the predictions are often in the right order of magnitude, which suggests that our 2024 forecasts are roughly reasonable, though they should be interpreted with caution. More data or a more advanced model could help improve accuracy.
